```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/2025-VeloCityX-Expanded-Fan-Engagement-Data.csv'
data = pd.read_csv(file_path)

# Check the number of missing values before removing rows
print("Missing values before removal:")
print(data.isnull().sum())

# Drop rows with missing values
data_cleaned = data.dropna()

# Check the number of missing values after removal
print("\nMissing values after removal:")
print(data_cleaned.isnull().sum())

# Verify the dataset shape after dropping missing values
print("\nShape of the dataset after removing missing rows:", data_cleaned.shape)

# Check for duplicate rows
duplicates = data_cleaned.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Remove duplicate rows if any
if duplicates > 0:
    data_cleaned = data_cleaned.drop_duplicates()
    print(f"\nDuplicates removed. New dataset shape: {data_cleaned.shape}")
else:
    print("No duplicates found.")

# Display the first few rows of the cleaned dataset
print("\nFirst few rows of the cleaned dataset:")
print(data_cleaned.head())

# Basic statistics on numerical columns
print("\nSummary statistics for numerical columns:")
print(data_cleaned.describe())

# Count of users who purchased virtual merchandise
merchandise_purchases = data_cleaned['Virtual Merchandise Purchases'].value_counts()
print("\nVirtual Merchandise Purchases distribution:")
print(merchandise_purchases)

# Plot the distribution of merchandise purchases
plt.figure(figsize=(8,6))
sns.countplot(x='Virtual Merchandise Purchases', data=data_cleaned)
plt.title('Distribution of Virtual Merchandise Purchases')
plt.show()

# Visualizing user engagement in key areas (Time on Live 360)
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Time on Live 360 (mins)'], kde=True)
plt.title('Distribution of Time on Live 360 (mins)')
plt.show()

# Additional visualization for Predictive Accuracy (%)
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Predictive Accuracy (%)'], bins=20, kde=False)
plt.title('Predictive Accuracy (%)')
plt.show()

# Additional visualization for Fan Challenges Completed
plt.figure(figsize=(10, 6))
sns.histplot(data_cleaned['Fan Challenges Completed'], bins=20, kde=False)
plt.title('Fan Challenges Completed')
plt.show()
```

Fan Challenge: "Ultimate Streak Challenge"
Overview:
The "Ultimate Streak Challenge" is designed to increase user engagement and incentivize continuous participation in predictions during multiple races or race events. The goal is to reward users for correctly predicting race outcomes over several consecutive events (building a streak). The longer the streak, the higher the potential rewards in terms of both engagement points and exclusive virtual merchandise.

Challenge Mechanics:

Streak Creation:
Users participate by predicting outcomes for various races in a series (e.g., race winners, fastest lap, most pit stops, etc.).
For each correct prediction, they build a streak (e.g., 1 correct prediction = 1 streak point).
The streak resets if the user makes an incorrect prediction.

Milestone Rewards:
Rewards are unlocked at streak milestones (e.g., 3, 5, 10 correct predictions).
Rewards include:
Exclusive Virtual Merchandise (e.g., race-themed skins, driver helmets, or team jerseys).
Bonus Fan Points that increase their leaderboard rank.
Discounts on Virtual Merchandise for longer streaks, encouraging users to make purchases.

Leaderboard:
A public leaderboard will showcase users with the longest streaks, motivating competition and engagement.
Users who rank on the leaderboard for the longest streaks will receive special virtual prizes at the end of the race season.
Incentives to Buy Virtual Merchandise:

Users can "boost" their chances of winning by purchasing exclusive prediction aids or unlocking hints through virtual merchandise. These boosts would help users make more accurate predictions, indirectly driving merchandise sales.

Predicted Outcomes for Engagement:

Increased Participation:
By creating a challenge that rewards consistent engagement, users will be incentivized to return for each race event. A similar concept to gamified loyalty programs, this will increase users time spent on the app and their likelihood of interacting with other features (like Live 360 or chat).
The streak element adds excitement and encourages users to avoid missing races in the series, boosting participation across events.

Higher Interaction in Fan Challenges:
Users will become more motivated to engage in prediction challenges as streak rewards are tied to accuracy. This can lead to increased participation in fan challenges, directly correlating to more engagement and longer app usage per session.

Increased Leaderboard Competitiveness:
Having a streak-based leaderboard will foster a sense of competition among users, motivating more predictions and increasing overall interaction.
